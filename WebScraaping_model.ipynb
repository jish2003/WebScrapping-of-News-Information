```python
import pandas as pd
import numpy as np
import requests
import cssutils
from bs4 import BeautifulSoup
from time import sleep
from random import randint
```

```python
title = []
image = []
fact = []
new_urls = []
```

```python
pages = np.arange(60,74)
```

```python
for pg in pages:
    page = requests.get("https://hindi.opindia.com/category/fact-check/page/"+str(pg)+"/")
    soup = BeautifulSoup(page.text, 'html.parser')
    news_data_main = soup.find('div', attrs = {'class': 'td_block_inner tdb-block-inner td-fix-index'})
    sleep(randint(2,8))
    news_data = news_data_main.findAll('div', attrs = {'class': 'tdb_module_loop td_module_wrap td-animation-stack'})
    #  div_style = news_data.find('span', attrs = {'style': 'background-image'})
    #  style = cssutils.parseStyle(div_style)
    for store in news_data:
        name = store.h3.a.text
        title.append(name)

        new_url = store.h3.a['href']
        new_urls.append(new_url)

        fact.append("False")

i = 1
for urls in new_urls:
    go_to = requests.get(urls)
    soup2 = BeautifulSoup(go_to.text, 'html.parser')
        
    image_data_main = soup2.find('div', attrs = {'class': 'td_block_wrap tdb_single_featured_image tdi_130 tdb-content-horiz-left td-pb-border-top td_block_template_1'})
    sleep(randint(2,8))
    image_url = image_data_main.img['src']
    #image.append(image_url)
    
    file_name = 'img-opindia' + str(i)
    i = i + 1
    sleep(randint(2,5))
    tempRes = requests.get(image_url)
    img_file = open(file_name + '.jpeg', 'wb')
    img_file.write(tempRes.content)
    img_file.close()
    image.append(file_name)
        
        

    

# dataset1 = pd.DataFrame({"Title": title, "Fact": fact, "Urls": image})
        
        
    
```
